```python
with open("dataset/1000.txt", 'r') as f:
    lines = f.readlines()
edges = []
for line in lines:
    x, y, z = line.strip().split()
    edges.append((int(x), int(y)))
```

```python
with open("dataset/1000input.txt", 'r') as f:
    lines = f.readlines()
points = []
for line in lines:
    x, y = line.strip().split()
    points.append((float(x), float(y)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as data
%matplotlib inline
sns.set_context('poster')
sns.set_style('white')
sns.set_color_codes()
plot_kwds = {'alpha' : 0.25}
```

```python
x = []
for u, v in edges:
    x1, y1 = points[u]
    x2, y2 = points[v]
    x.append([x1, x2])
    x.append([y1, y2])
plt.plot(*x, marker = '.', color='b', **plot_kwds);
```
